```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # graficos

from sklearn import datasets

%matplotlib inline

# pegando os dados da tabela de vinhos
vinhos = datasets.load_wine()
# lendo os dados e as separando em dados e nome de cada coluna
df = pd.DataFrame(vinhos['data'], columns=vinhos['feature_names'])  
# renderizar com as seguintes dimenções altura e largura
df.hist( figsize=(15,10) );
```

```
import seaborn as sns

df['labels'] = vinhos['target']
#criando graficos do tipo de bolinhas
sns.pairplot( df, hue='labels', vars=df.drop(columns='labels') )
```

utilizando o K-means

```
# utilização do K-means
from sklearn.cluster import KMeans

X = df.drop(columns='labels')

intra_cluster = []
for k in range(2, 10):

  #numero de clusters é o numero que estiver na iteração(K)
  model = KMeans(n_clusters=k)
  # 
  model = model.fit(X)
  #
  intra_cluster.append(model.inertia_)

plt.figure(figsize=(10,5))#tamanho do grafico
# grafico montando x e y com os dados do range do cluster e o valor entre os cluster
plt.plot(range(2,10), intra_cluster) 
```

```
# Graficos

X['labels'] = model.labels_
sns.pairplot( X, hue='labels', vars=X.drop(columns='labels') )
```
