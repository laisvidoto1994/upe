```
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
import seaborn as sns

data = sns.load_dataset('iris').drop(columns='species').values

Z = linkage(data)

plt.figure(figsize=(15, 8))

dendrogram(Z)  

plt.show()
```

```
from scipy.cluster import hierarchy
import seaborn as sns

dados = sns.load_dataset('iris').drop(columns='species')

Z = hierarchy.linkage(dados, 'average')

plt.figure( figsize=(15, 6) )
 
dn = hierarchy.dendrogram(Z)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

%matplotlib inline

# pegando os dados da tabela de iris
iris = datasets.load_iris()
# lendo os dados e as separando em dados e nome de cada coluna
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
# renderizar com as seguintes dimenções altura e largura
df.hist(figsize=(15,10));
```

```
df['labels'] = iris['target']
sns.pairplot(df, hue='labels', vars=df.drop(columns='labels'))
```

```
from sklearn.cluster import KMeans

X = df.drop(columns='labels')

intra_cluster = []
for k in range(2, 4):

  model = KMeans(n_clusters=k)
  model = model.fit(X)
  intra_cluster.append(model.inertia_)

plt.figure(figsize=(10,5))
plt.plot(range(2,4), intra_cluster)
```

```
X['labels'] = model.labels_
sns.pairplot(X, hue='labels', vars=X.drop(columns='labels'))
```
